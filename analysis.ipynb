```python
import pandas_datareader as pdr
import pandas as pd
import numpy as np
import datetime as dt
```

```python
start = dt.datetime(2000, 1, 1)
end = dt.datetime(2022, 11, 1)
usdcad = pdr.DataReader("DEXCAUS", 'fred', start, end)
```

```python
var_exp_dict = {}

var_exp_df = pd.DataFrame()

var_exp = ['dhhngsp','dcoilwtico','dcoilbrenteu', 'will5000prfc','dpropanembtx','dhoilnyh','willresipr',
           'vixcls','gvzcls','ovxcls','vxvcls', 'dexuseu', 'dexusuk', 'dexchus', 'dexjpus', 'dexkous',
           'dexmxus', 'dexinus', 'dexvzus', 'dexbzus', 'dexszus', "DTWEXEMEGS", "USRECD", "EURORECD", "JPNRECDP"
          ]

for i in var_exp:
    # téléchargement sur la fed américaine à l'aide de pandas datareader
    var_exp_dict[i] = pdr.DataReader(i.upper(), 'fred', start, end)
    if var_exp_df.empty:
        var_exp_df = var_exp_dict[i]
    else:
        var_exp_df = var_exp_df.join(var_exp_dict[i])
    var_exp_dict[i].ffill(inplace=True)
```

```python
var_exp_df = usdcad
for i in var_exp:
    # téléchargement sur la fed américaine à l'aide de pandas datareader
    var_exp_dict_2[i].pct_change()
    var_exp_dict_2[i]= var_exp_dict_2[i]
    var_exp_df[i] = var_exp_dict_2[i][i.upper()]
    #var_exp_dict[i] = np.array(var_exp_dict[i])  
```

```python
var_exp_df.ffill(inplace=True)
var_exp_df.dropna(inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
var_exp_df.corr()
```

```python
var_dep_df = var_exp_df.pop('DEXCAUS')
```

```python
start = var_exp_df.index[0].date()
end = var_exp_df.index[-1].date()
print(start, end)
```

```python
var_dep = np.array(var_dep_df)
var_exp = np.array(var_exp_df)
```

```python
X_train, x_test, y_train, y_test =train_test_split(var_exp, var_dep, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
coeffs = []

for i, coeff in enumerate(model.coef_[1:]):
    coeffs += [[var_exp_df.columns[i], coeff]]
```

```python
coeffs_df = pd.DataFrame(coeffs)
coeffs_df.columns = ['Variable', 'Coefficient']
coeffs_df.sort_values(by='Coefficient', ascending=True, inplace=True)
coeffs_df.set_index('Variable', inplace=True)
```
